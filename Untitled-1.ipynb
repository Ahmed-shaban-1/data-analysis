```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"A:\trayinggggggggg\data\Instagram data.csv", encoding="latin1")
print(df.head())
```

```python
df.plot(kind='scatter', x='Impressions', y='Likes', title='Likes vs Impressions')
plt.show()
print(df[['Impressions', 'Likes']].corr())
```

```python
sources = ['From Home', 'From Hashtags', 'From Explore', 'From Other']
for src in sources:
    df[f'{src}_to_follow'] = df['Follows'] / df[src]
print(df[[f'{s}_to_follow' for s in sources]].mean().sort_values(ascending=False))
```

```python
df['Hashtag_Count'] = df['Hashtags'].fillna('').apply(lambda x: len(x.split('#')) - 1)
df.plot(kind='scatter', x='Hashtag_Count', y='Impressions', title='Hashtag Count vs Impressions')
plt.show()
print(df[['Hashtag_Count', 'Impressions']].corr())
```

```python
top_save = df.loc[df['Saves'].idxmax()]
print(top_save[['Caption', 'Saves', 'Impressions', 'Likes']])
```

```python
print(df[['Comments', 'Likes']].corr())
df.plot(kind='scatter', x='Comments', y='Likes', title='Comments vs Likes')
plt.show()
```

```python
top_profiles = df.sort_values(by='Profile Visits', ascending=False).head(5)
print(top_profiles[['Caption', 'Profile Visits']])
```

```python
df['Follow_Rate'] = df['Follows'] / df['Impressions']
print(df[['Caption', 'Follow_Rate']].sort_values(by='Follow_Rate', ascending=False).head(5))
```

```python
df['In_Explore'] = df['From Explore'] > 0
print(df.groupby('In_Explore')[['Likes', 'Impressions', 'Follows']].mean())
```

```python
print(df[['Shares', 'Likes', 'Comments', 'Saves']].corr())
```

```python
df[sources].mean().plot(kind='pie', autopct='%1.1f%%', title='Average Reach Sources')
plt.ylabel('')
plt.show()
```

```python
tag = "#datascience"
df['Has_tag'] = df['Hashtags'].fillna('').str.contains(tag, case=False)
print(df.groupby('Has_tag')[['Impressions', 'Likes', 'Follows']].mean())
```

```python
df['Caption_Length'] = df['Caption'].apply(lambda x: len(str(x)))
df.plot(kind='scatter', x='Caption_Length', y='Impressions', title='Caption Length vs Impressions')
plt.show()
print(df[['Caption_Length', 'Impressions']].corr())
```

```python
corr = df[['Follows', 'Impressions', 'Likes', 'Comments', 'Shares', 'Saves']].corr()['Follows'].sort_values(ascending=False)
print(corr)
```
